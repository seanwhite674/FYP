```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Mean and covariance for multivariate normal distribution
mean, cov = [0., 0.], [(1., -0.6), (-0.6, 1.)]

# Generate random data
data = np.random.multivariate_normal(mean, cov, 1000)

print(data)

# Create a DataFrame
df = pd.DataFrame(data, columns=["x1", "x2"])

print(df)

# Create the jointplot using keyword arguments for x, y, and data
g = sns.jointplot(x="x1", y="x2", data=df, kind="kde")

# Plot additional scatter points on top of the KDE plot
g.plot_joint(plt.scatter, c="g", s=30, linewidth=1, marker="+")

# Set the axis labels
g.set_axis_labels("$x1$", "$x2$")

# Show the plot
plt.show()
```
